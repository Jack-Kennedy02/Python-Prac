```python
import pandas as pd
df = pd.read_csv("cat_breeds_dirty.csv", delimiter=';')

```

```python
df.isnull().sum()      # Count of missing values per column
```

```python
df_clean = df.dropna(subset=['Breed'])
df_clean.isnull().sum()
```

```python
df_clean = df_clean[df_clean['Age_in_years'] >= 0]
df_clean = df_clean[df_clean['Age_in_months'] >= 0]

df_clean.describe()
```

```python
df_clean = df_clean.dropna()
df_clean.isnull().sum()
```

```python
df_pivot = df_clean.pivot_table(values='Weight', index='Preferred_food', aggfunc='mean')
df_pivot.head()
# There isn't a difference between preferred food and weight.
```
